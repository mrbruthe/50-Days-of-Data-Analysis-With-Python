2. Using NumPy, write a code that returns all the numbers in 
the array that are greater than 25. 

```python
# Getting indexes of numbers over 25
arr_g25 = np.where(arr > 25)
arr_g25

# Using the index to get numbers over 25
print(arr[arr_g25])
```

3. Here is some data below:    
days = ["mon", "tue",  "wed",  "thu", "Fri"]     
hours_worked = [ 8, 8, 10, 11, 7]     
Using Boolean indexing of NumPy arrays, return the days 
of the week where hours worked were 8. 

```python
days = ["mon", "tue",  "wed",  "thu", "Fri"]     
hours_worked = [ 8, 8, 10, 11, 7]  

# Creating arrays
days = np.array(days)
hours_worked = np.array(hours_worked)

# Using Boolean to get the 8 hours days.
days_8hrs = days[hours_worked == 8]
days_8hrs
```

4. Write another code (using Boolean indexing) to return the 
day of the week with the highest number of hours worked. 

```python
# finding the day of the week with highest number of hours worked

max_hours = np.argmax(hours_worked)
print(f"Day with the max hours worked is, {days[max_hours]}")
```

5. Write another code to return all the days of the week where 
over 9 hours of work were achieved. 

```python
days_9 = days[hours_worked > 9]
days_9
```
