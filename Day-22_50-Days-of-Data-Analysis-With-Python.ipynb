1. Using pandas create a DataFrame with the two lists above. 
Write a code that counts how many missing values are in 
each column. 

```python
import pandas as pd
import numpy as np

names = ["Carol", "Kate", "Jane", "Kuda", "Tito", "Kuku"]
age = [23, np.nan, 34, 56, np.nan, 44]


df = pd.DataFrame({"names":names, "age":age})
df
```

```python
# Checking for missing values
df.isna().sum()
```

2.  Create a copy of the DataFrame in question 1 and drop all 
the missing values. 

```python
# Creating a copy of the DataFrame
df_copy = df.copy()

df_copy = df_copy.dropna()
df_copy
```

3. You decided you donâ€™t want to drop the missing numbers, 
but replace them with the mean of the numbers in the 
column. Create another copy of the DataFrame (from the 
original DataFrame in question 1) and fill in the missing 
values with the mean of the column. 

```python
df2 = df.copy()

df2 = df2.fillna(df2["age"].mean(), inplace = False)
df2
```

4. You have come across some more data: 

```python
gender_values = ["F","F","F","M","M","M"]
```

Update your DataFrame (question 3) with the gender 
column using the gender_values list above. Use the melt() 
method to reshape the DataFrame and create a new 
DataFrame with three columns: the "names" column, the 
"gender" and "age"  columns will be combined, and their 
values will be in the values column. This will change the 
DataFrame from a wide format to a long format.  

```python
df2["gender"] = gender_values
df2
```

```python
melted_df = df2.melt(id_vars=["names"],
                     value_name="values",
                     var_name="ages_gender")

melted_df
```

5. Using the pandas where() method, write a code to return a 
DataFrame of names and gender values only from the 
"melted" DataFrame (question 4). Your DataFrame should 
not have NaN values. Save this to the new variable. 

```python
# names and gender values from the melted DataFrame
names_gender = melted_df.where(melted_df["ages_gender"]=="gender").dropna()
names_gender
```

6. Convert the column header of the DataFrame in question 5 
to uppercase letters. 

```python
# converting the column headers into uppercase letters 
names_gender.columns = names_gender.columns.str.upper()
names_gender
```
