# Day 3: Generating Random Arrays

1. create an array of random numbers 
between 0 and 1. The shape of the array must be (3, 4). 

```python
import numpy as np

# Creating random number betwen 0 to 1 
rng = np.random.default_rng()
arr1 = rng.random((3, 4))
arr1
```

Numbers between 0 to 1 means floats, and in creating floats we use rng.random 

2. Create an array of random integers between 0 and 10. The 
shape of the array must be (3, 4). 

```python
# Creating random integers between 0 to 10 

arr_itg = np.random.default_rng()
arr_itg = rng.integers(low=1, high=11, size=(3,4))
arr_itg
```

```python
import matplotlib.pyplot as plt 

# Instantiating a generator and setting seed
rng = np.random.default_rng(seed=42)
arr2 = rng.standard_normal(size=(1000))

# plotting a hist plot
plt.hist(arr2, bins=50)
plt.title("Standard Normal Distribution")
plt.show()
```

4. arr = [ "Orange", "Apple", "Pear"] 
Using random.choice() from NumPy, generate an array 
from the list above. The shape of the array must be (3, 4). 

```python
arr3 = [ "Orange", "Apple", "Pear"]

# constructing a random generator
rng = np.random.default_rng(seed=42)
arr4 = rng.choice(arr3, size=(3, 4))
arr4
```
