## Day 20: Explore Data with Pandas and Matplotlib 

```python
products = ["sugar", "Salt", "oil", "diapers", "rice"]

costs = [2450, 1989, 6745, 9807, 8743]

sales = [27908, 4508, 6734, 9976, 9000]
```

1. Create a pandas DataFrame from the lists above. Find the 
most profitable product. Using Matplotlib, create a bar 
plot to visualize product profitability. 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({"Products": products, "Costs": costs, "Sales": sales})
df
```

```python
# calculate the profit for each product
df["Profit"] = df["Sales"] - df["Costs"]

# most profitable product index
most_profitable_index = df["Profit"].idxmax()

# finding the corresponding product with the index 
most_profitable_product = df.loc[most_profitable_index, "Products"]
print(f"The most profitable product is {most_profitable_product}")
```

```python
plt.figure(figsize=(8,6))
plt.bar(df["Products"], df["Profit"])
plt.xlabel("Products")
plt.ylabel("Profit")
plt.title("Product Profit", fontsize=20)
plt.show()
```

2. Using just pandas, find the least profitable product. 


```python
least_profitable_index = df["Profit"].idxmin()
least_profitable_product = df.loc[least_profitable_index, "Products"]

print(f"The least profitable product is {least_profitable_product}")
```

3. What is the difference in profit between the most 
profitable item and the least profitable item? 

```python
max_min_profit_difference = df.loc[most_profitable_index, "Profit"] - df.loc[least_profitable_index, "Profit"]

print(f"The difference between the most and least profiting product is {max_min_profit_difference}")

```

4. Using pandas, and Matplotlib, create a line plot of the 
costs and profits of all products. 

```python

df.plot(x="Products", y=["Costs", "Profit"],
        kind="line",
        figsize=(6, 4))

plt.xlabel("Products")
plt.ylabel("Cost and Profit")
plt.title("Cost and Profit Analysis", fontsize=20)
plt.show()
```

5. Using pandas, calculate the average cost per product. 


```python
average_cost = df["Costs"].mean()
print("The average costs is", average_cost)
```

6. Using pandas calculate the average profit per product. 


```python
average_profit = df["Profit"].mean()
print("The average profit is", average_profit)
```
