1. Import the sports dataset and write code to check the first 
five rows to confirm that it has loaded properly. Now, using 
the pandas items() method, return all the names of the 
columns that have values of the "object" data type. 


```python
import pandas as pd
df = pd.read_csv("sports_data.csv")
df.head()
```

```python
columns_names = []

# Looping over the DataFrame
for (columns, values) in df.items():
    if values.dtypes == 'object':
        columns_names.append(columns)

print(columns_names)
```

2. Using a pandas method, check the memory size of each 
column using the df.memory_usage() method. Set the deep 
parameter to True. What is the total memory usage of the 
DataFrame in kilobytes? 


```python
df.memory_usage(deep = True).sum()/1024
```

3. Using pandas convert all the columns with the "object" data 
type to the "categorical" data type. Create a new variable for

```python
# Changing the data type from object to category
df2 = df.astype({"Name": "category", "Sport":"category", "Team":"category"})
df2.dtypes
```

```python
df2.memory_usage(deep = True).sum()/1024
```

4. Using the pandas method, rename the "Yellow" column to 
"Yellow Cards." 


```python
df2 = df2.rename(columns={'Yellow':'Yellow Cards'})
df2.head()
```

5. Using the pandas query() method, create a subset 
DataFrame of only players that got over six fouls. Reset 
your index (set it back to the default 0, 1, 2, etc.) and ensure that you remove the index as a column. 


```python
# Using query to filter data by fouls 
df3 = df2.query("Fouls > 6").reset_index(drop=True)
df3
```
