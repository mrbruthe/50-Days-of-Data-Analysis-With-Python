1. Import the medical_data dataset and view the first 5 rows.

```python
import pandas as pd

df = pd.read_csv("medical_data.csv")
df.head(5)
```

2. What are the IQR ranges of the "BMI" column? 


```python
df.quantile

lower_limit = df['BMI'].quantile(q = 0.25)
upper_limit = df['BMI'].quantile(q = 0.75)


print(f'THe IQR range is {lower_limit:.2f} to {upper_limit:.2f}')
```

3. You must have noticed that there are two columns with 
names. As part of feature engineering, you are required to 
create a copy of your DataFrame (question 1). Combine the 
"First Name" column and the "Last Name" column into one 
column called "Names." 


```python
df2 = df.copy()
df2["Names"] = df2["First Name"].str.cat(df2["Last Name"], sep='')
df2.head()
```

4. Now that you have combined the two columns, drop 
columns "First Name" and "Last Name" from the 
DataFrame and change the order of the columns to: 
[Names, First Letter, BMI, Gender, Condition]. 


```python
# Drop the two columns
df2.drop(columns=(["First Name", "Last Name"]), axis=1, inplace=True)

# using iloc to change columns order
df2 = df2.iloc[:, [4, 0, 3, 2, 1]]
df2.head()
```

5. The "Gender" column is combined with the age. Your task 
is to create a new column for age by separating age from the 
"Gender" column. Once you do that, make the "Condition" 
column the last column of the DataFrame. Your columns' 
order must be: [Name, First Letter, BMI, Gender, Age, 
Condition]. 

```python
# Creating two columns from the Gender column using the split method
df2[['Gender', 'Age']] = df2.Gender.str.split(" ", expand=True)

# using iloc to change columns order
df2 = df2.iloc[:, [0, 1, 2, 3, 5, 4]]
df2.tail()
```
