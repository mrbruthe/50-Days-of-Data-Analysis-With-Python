1. Create two arrays of random integers between 10 and 20. 
The shape of the array must be (2, 3) for the first array 
and (1, 3) for the second array. Check the shape of the 
arrays. Ensure that the results are reproducible. 

```python
import numpy as np

# Creating random array
rng = np.random.default_rng(seed=42)
arr1 = rng.integers(low=10, high=20, size=(2,3))
print(arr1)


np.shape(arr1)
```

```python
arr2 = rng.integers(low=10, high=20, size=(1, 3))
print(arr2)

np.shape(arr2)

```
