## Day 14: Analyze Data with NumPy Part - 1 

```python
names   = ["Kelly", "Ben", "Jack", "Muhammad", "Jose", "Liz"]
grades  = [45, 65, 75, 35, 85, 40]
classes = ["a", "C", "e", "g", "e", "f"]
```

1. Using NumPy, write code that returns all the names and 
the  number of  students from class "e" who got over 50 
marks. 

```python
import numpy as np

arr = np.array([names, grades, classes])
arr
```

```python
# Slicing names from array 
names = arr[0:1,0:]

# Slicing grades from array and converting array to int 
grades = arr[1:2,0:].astype(int)

# Slicing classes from array 
classes = arr[2:3, 0:]

# find the index of students in class 'e' who got over 50 
student_index = np.where((classes == "e") & (grades > 50))

# number of students
over_50_names = len(names[student_index])

# names of students
class_e_students_over_50 = names[student_index]

print(f"Student from class 'e' with over 50 are: {class_e_students_over_50}")
print(f"Number of Students who got over 50 marks in class 'e' are: {over_50_names}")


```

2. Your next task is to write a code that returns all the students 
who got less than 75 marks. 


```python
below_75_index = np.where(grades < 75)

names_75 = names[below_75_index]
names_75
```

3.  Write code that returns how many marks a student named 
Liz got. 

```python
# Find the index of the student named "Liz"
index = np.where(names == "Liz")


# Use the index to get the student's marks
liz_marks = grades[index][0]
print(f"Liz got {liz_marks} marks")
```

4.  Which student got the highest marks? 

```python
maximum = np.argmax(grades)

name = names[:, maximum]
print(f'The student who got the highest marks is, {name[0]}')
```

5. What is the longest name in the array, and what is its 
index?  

```python
longest_word = np.char.str_len(names)

index_longest_word = np.argmax(longest_word)

longest_word = arr[:, index_longest_word]

print(f"The longest name is sitting at index: {index_longest_word}")
print(f"The longest name is : {longest_word[0]}")
```
