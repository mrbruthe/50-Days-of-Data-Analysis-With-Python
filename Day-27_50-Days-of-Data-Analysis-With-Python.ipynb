```python
import pandas as pd

df = pd.read_csv("retail_shop_data_reprocessed.csv")
df.head(21)
```

1. Using .loc attribute, calculate the profit of the sunglasses. 


```python
# Find the sales value
sales_sunglasses = df.loc[df["Product Name"]=="Sunglasses", "Revenue"]

# Find the cost value 
costs_sunglasses = df.loc[df["Product Name"]=="Sunglasses", "Costs"]

# Calculate the difference to find profit 
profit_sunglasses = sales_sunglasses - costs_sunglasses

print(f"The profit of Sunglasses is ${profit_sunglasses.iloc[0]}")
```

2. Using .loc attribute, calculate the profit of hoodies. 


```python
# Find the sales value
sales_hoodie = df.loc[df["Product Name"]=="Hoodie", "Revenue"]

# Find the cost value 
costs_hoodie = df.loc[df["Product Name"]=="Hoodie", "Costs"]

# Calculate the difference to find profit 
profit_hoodie = sales_hoodie - costs_hoodie

print(f"The profit of Hoodie is ${profit_hoodie.iloc[0]}")
```

3. Using .iloc attribute, return the names of the two least 
profitable products. 

```python
# Sort products by profit in ascneding order

df.sort_values("Profit", inplace=True)

# Use iloc to find 2 least profitable products.
least_profitable_products = df.iloc[:2,3]
least_profitable_products
```

5. Using the Seaborn library, create a scatter plot to visualize 
the relationship between sales and costs for each product. 
Is there any noticeable correlation?

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(x=df["Revenue"], y=df["Costs"], data=df)
plt.title("Costs versus Profit", fontsize=20)
plt.show()
```
