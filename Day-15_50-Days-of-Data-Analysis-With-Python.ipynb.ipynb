## Day 15: Analyse Data with NumPy Part - 2 


1. Your task is to import the CSV file using NumPyâ€™s 
genfromtxt() function.   
a. Transpose the array.     
b. Using slicing, create three arrays: an array of all the 
names in the column, an array of the age column, 
and an array of the gender (sex) column. 

```python
import numpy as np
arr = np.genfromtxt("names_age_sex_data.csv", delimiter=",", dtype ='U7')
arr
```

```python
transposed_arr = arr.transpose()
transposed_arr
```

```python
# slicing names
names = transposed_arr[0:1,1:]
names
```

```python
# slicing age
age = transposed_arr[2:3,1:]
age
```

```python
# slicing sex
sex = transposed_arr[1:2,1:]
sex
```

2. Using NumPy, write code that returns all the names of 
people who are 56 years old. 

```python
age_56 = names[age=='56']
age_56
```

3. How many people are aged 44 or under? 


```python
age_44_below = names[age<='44']

# using the shape attribute to find people under 44
age_44_below_num = np.shape(age_44_below)
print("The number of people aged 44 and under is", age_44_below_num[0])
```

4. Write another code snippet to return an array of all the 
males in the dataset. How many males are in the dataset? 

```python
all_male = names[sex=="Male"]
all_male_num = len(all_male)

print(all_male)
print("The number of males is:", all_male_num)
```

5. Calculate the average age of all the females in the table. 


```python
avg_female_age = age[sex=="Female"].astype(int).mean()

print("The average females age is ", avg_female_age)
```

6. Calculate the average age of people named Olivia and 
Kate.  

```python
age = age.astype(int)
olivia_age = age[names == 'Olivia']
kate_age = age[names == 'Kate']
avg_female_age_n = np.array([olivia_age, kate_age]).mean()
print(f"Average age ofpersons named Olivia and Kate is: {avg_female_age_n:.2f}")
```
