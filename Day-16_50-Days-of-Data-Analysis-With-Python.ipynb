```python
import pandas as pd

# Creating a panda series 
colors_series = pd.Series(list1)
colors_series
```

```python
colors = ["red", "white", "black"]

# Checking if the colors in the list are in series
is_in_series = colors_series.isin(colors)
is_in_series
```
