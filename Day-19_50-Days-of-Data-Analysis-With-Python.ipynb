```python
names = ["Joe", "Phil", "Ken", "Jos", "Luke"]
miles_run = [120, 80, 100, 90, 85]
time_in_hrs = [40, 38, 45, 50, 50]
```

1. Create a DataFrame from the lists above using pandas 


```python
import pandas as pd

df = pd.DataFrame({"Names": names, "Miles run": miles_run, "Time in hrs":time_in_hrs})
df
```

2. More information has become available below. Create a 
new DataFrame with this information. Merge it to the 
DataFrame above using pandas concat() function. 

```python
# New information
names = ["Joe", "Phil", "Ken", "Jos", "Luke"]
Age = [45, 28, 21, 55, 62]
gender = ["M", "M", "F", "F", "M"]
```

```python
df_new = pd.DataFrame({"Names": names, "Age": Age, "gender": gender})
df_new
```

```python
df = pd.concat([df, df_new.drop(columns="Names")], axis=1)
df
```

3. Shuffling DataFrames can randomize the data. Write code 
to reshuffle the index of the DataFrame. 

```python
# Reshuffling the dataset
df = df.sample(frac=1)
df
```

4. Explain why it may be important to reshuffle the index of 
the data for machine learning. 

Reshuffling the index of data is important in machine learning because it helps prevent order bias and ensures that models learn general patterns rather than memorizing the sequence of the data. reshuffling helps maintain the integrity of the learning process by ensuring randomness and reducing bias.

5. Using the pandas melt() function, reshape your DataFrame
