```python
import pandas as pd

df = pd.read_csv("spare_parts_expanded.csv")
df.dtypes
```

```python
df.isnull().sum()
```

```python
df["spare_parts"].duplicated().sum()
```

```python
# Remove comma from the total revenue column
df["total_revenue"] = df["total_revenue"].str.replace(",", "", regex=False)

# Remove $ from the total_revenue column
df["total_revenue"] = df["total_revenue"].str.replace("$", "", regex=False).astype(int)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

plt.hist(df["total_revenue"], bin=10)
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.title("Distribution of Total Sales", fontsize = 12)
```
