## Day 35: Runners And Income Data  Analysis 


In this challenge, you will preprocess and analyze runners 
data. You will import the runners_and_income_data file. 


1. Load the dataset using pandas and view the first two rows 
of the dataset. You want to know how many rows and 
columns are in the DataFrame; write a code to check the 
number of rows and columns. 


```python
import pandas as pd

df = pd.read_csv("runners_and_income_data.csv")
df.head()
```

```python
print(f"The number of rows are {df.shape[0]}")
print(f"The number of columns are {df.shape[1]}")
```

2. Write a line of code to check the total NaN values in the 
DataFrame. Drop rows that have all NaN values. Replace 
the remaining NaN values with the value 00.0 (use 
pandas). 


```python
print(f"The total number of NaN rows are {df.isna().sum().sum()}")
```

```python
# Drop rows with all NaN values
df.dropna(axis=0, how='all', inplace = True)
```

```python
# Replace NaN values with 0.00
df.fillna(value=0.00, inplace=True)
```

```python
print(f"Number of NaN values: {df.isna().sum().sum()}")
df.head()
```

3. What does the describe() method do? Use describe to 
return the mean of the "Miles Run" column. 


```python
# describe method returns the statistical summary of the DataFrame
df.describe()["Miles Run"]["mean"]
```

4. Using the np.sum, what is the total number of miles run by 
all the names in the DataFrame? 


```python
import numpy as np
print(f"Total miles run: {np.sum(df['Miles Run'])}")
```

5. How many people have an income of $50000? Return a list 
of names. 


```python
# Using query and iloc to filter the DataFrame for names 
names_of_5000_income = df.query("Income== 50000").iloc[0:,0]
print(list(names_of_5000_income))
```
