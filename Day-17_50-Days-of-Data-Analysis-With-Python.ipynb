4. Create a DataFrame using the data below:

    names = ["John", "Mary", "Peter"]    
    age = [27, 34, 47]    
    sex = ["male", "female", "female"]
      
    Using the .loc attribute, access the age of John from the  DataFrame. 

```python
names = ["John", "Mary", "Peter"]
age = [27, 34, 47]
sex = ["male", "female", "female"]

df = pd.DataFrame({"names" : names, "age": age, "sex":sex})
john_age = df.loc[0, "age"] 
john_age
```

5. Using the .loc attribute change the gender of Peter from 
"female" to "male." 

```python
# Chaging Peter's gender using loc
ds = df.loc[2, "sex"] = "male"
df
```

6. Write code to extract a subset of the DataFrame containing 
only male individuals. Write another code to retrieve the 
name and age of the oldest male from the modified 
DataFrame. 

```python
# using query method to filter the DataFrame
male = "male"
males = df.query('sex == @male')
males
```

```python
# Sorting the DataFrame and returning the oldest male
oldest = males.sort_values('age', axis=0, ascending=False).max()

print(f"The oldest person is {oldest["names"]} and he is {oldest["age"]} years old")
```
