# Seaborn
It is also an open-source library. 
Seaborn is built on top of the Matplotlib library; as such, it 
integrates well with the functions of Matplotlib as well as 
the pandas library.

The below is the data to be used in this section.

```python
import pandas as pd

products = ['Cars', 'Beds', 'Phones', 'Shoes',
            'computers', 'clothes', 'Books', 'Boats',
            'Car oil', 'Scooter']

categories = ['Transportation', 'Household', 'Electronics',
              'Personal Items', 'Electronics', 'Personal Items',
              'Personal Items', 'Transportation', 'Transportation','Transportation']

sales = [2000, 1500, 1000, 1200, 3000, 1000, 1200, 1300, 1000, 2500]

costs = [1500, 1000, 700, 800, 2500, 700, 700, 1000, 900, 1100]

profit = [500, 500, 300, 400, 500, 300, 500, 300, 100, 1400]

# Creating a DataFrame from the four lists above
df = pd.DataFrame({"Products": products, "Categories": categories,
                   "Sales": sales, "Costs": costs, "Profit": profit})

df.head()
```

### histplot() 
shows the distribution of numerical data as a histogram, often with a KDE curve to highlight skewness or concentration. It’s ideal for spotting where most values lie and whether the data is symmetric or skewed.



```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Sales'], kde=True)
plt.title("Sales Values Distribution")
plt.show()
```

KDE stands for Kernel Density Estimate. It’s like a smooth version of a histogram.

A histogram counts how many data points fall into each bin — like stacking blocks.

A KDE curve draws a smooth line over those blocks to show the shape of the data — like tracing the outline of a mountain range.

### lineplot() 
visualizes trends between two continuous variables, helping detect patterns or correlations — even nonlinear ones — like how costs change with sales.



```python
# Creating a line plot
sns.lineplot(x=df['Sales'], y=df["Costs"], data=df)
plt.title("Sales and Costs Analysis")
plt.show()
```

From the plot, we can conclude that the relationship 
between sales and costs is nonlinear. Costs initially 
increase gradually with sales, but there are significant 
jumps at certain sales values.

### pairplot() 
displays pairwise relationships across multiple variables, combining scatter plots and histograms. It’s great for exploring correlations and seeing how categories (like product types) influence metrics.



```python
sns.pairplot(df, hue="Categories")
plt.title("Pair Plot of Products", y=3.10, x=-1, fontsize = 20)
plt.show()
```

This code uses Seaborn’s pairplot() to visualize pairwise relationships between all numeric columns in the DataFrame. The hue="Categories" argument colors the plots by product category, helping you see how different types of products relate across metrics like sales, costs, and profit. 

### regplot() 
combines a scatter plot with a regression line to show linear relationships between two numeric variables, such as how profit tends to increase with sales. It also makes outliers easy to spot.



```python
sns.regplot(x=df['Sales'], y=df["Profit"], data=df)
plt.title("Sales and Profit")
plt.show()
```

From the graph, we can generally conclude that there 
appears to be a positive linear relationship between sales 
and profit. As sales increase, profit tends to increase as 
well. 

### boxplot() 
summarizes the distribution of data across categories using quartiles, medians, and whiskers. It’s perfect for comparing groups and identifying outliers — like seeing which category yields the highest profit.

```python
sns.boxplot(x='Categories', y='Profit', data=df)
plt.title("Boxplot of Values by Category")
plt.show()
```

The line inside the box 
represents the median (the 50th percentile) of the data. 
The bottom part of the blue box represents the 25th 
quarter of data, and the top part of the blue box is the 75th 
quarter of the data in the category. The whiskers (the line 
extending lower and upper from the box) represent data 
points that lie within 1.5 IQRs of the lower and upper 
quartiles. Anything outside the whiskers is considered to 
be an outlier of the data. From this plot, we can see that 
the biggest junk of profit is from the transportation 
category.
