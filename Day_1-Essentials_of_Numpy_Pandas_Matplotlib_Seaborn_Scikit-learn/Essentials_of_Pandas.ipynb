## Pandas Series
A pandas Series is a one-dimensional array, a series has one column and an index(row lable)

```python
import pandas as pd

# Creating a Series 
fruits = pd.Series(["Orange", "Apple", "Mango"], name = "fruits")
fruits
```

**Series Index and Name**  
We can specify the values of the index using the index parameter.  The specified index must be the same length as  the items (data) in the Series and does not have to be unique (meaning you can pass the same index value for all 
the data).
 

```python
import pandas as pd

# Creating a Series 
fruits = pd.Series(["Orange", "Apple", "Mango"], name = "fruits", index = ['a', 'b', "c"])

fruits
```

**Changing Data Type**

This can be done using the astype() method.


```python
num_series = pd.Series([10, 20, 30, 40])
num_series
```

```python
# Changing data type to float
num_series = num_series.astype("float64")
num_series
```

## Creating a Pandas Data Frame
To create a Pandas DataFrame, pass a dictionary or list of lists to the DataFrame function. Column names are specified separately using the columns parameter.


```python

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 32, 18, 47],
    'Salary': [50000, 80000, 20000, 120000]
}

df = pd.DataFrame(data, columns=['Name', 'Age', 'Salary'])
df
```

We can also create a DataFrame from lists by passing them 
to the DataFrame function as a dictionary. See below: 

```python
names = ['Alice', 'Bob', 'Charlie', 'David']
age = [25, 32, 18, 47]
salary = [50000, 80000, 20000, 120000]

df = pd.DataFrame({"Names": names, "Age": age, "Salary": salary})
df
```

## Data Cleaning

### .dropna()
Removes rows or columns with missing (NaN)values.


```python
names = ['Alice', 'Bob', 'Charlie', 'David']
age = [25, None, 18, 47]
salary = [50000, 80000, None, 120000]

df = pd.DataFrame({"Names": names, "Age": age, "Salary": salary})
df
```

the NaN values represent the missing values, we will now drop all rows with missing values with the dropna() method.

```python
df.dropna()
```

```python
# To drop all columns with missing values
df.dropna(axis=1)
```

Note   
axis=0 refers rows    
axis=1 refers to columns     

### bfill and ffill()    

.bfill() → Backward fill: replaces missing values with the next valid value below.   

.ffill() → Forward fill: replaces missing values with the previous valid value above.    

```python
names = ['Alice', 'Bob', 'Charlie', 'David']
age = [25, None, 18, 47]
salary = [50000, 80000, None, 120000]

df = pd.DataFrame({"Names": names, "Age": age, "Salary": salary})

# Fill missing values in 'Age' using backward fill
df['Age'] = df['Age'].bfill()

# Fill missing values in 'Salary' using forward fill
df['Salary'] = df['Salary'].ffill()

df
```

### groupby()
Groups data by one or more columns and applies aggregate functions like sum(), mean(), or count().



```python
# A DataFrame of product sales
Products = ['Computers', 'Phones', 'Shoes', 'Computers', 'Phones']
Sales = [2500, 3000, 1400, 2100, 2800]

# Creating a DataFrame
df = pd.DataFrame({"Products": Products, "Sales": Sales})

# Viewing the first 2 rows
df.head(2)
```

```python
# Grouping prodiucts by the "Product" column and summing the sales
groupby_products = df.groupby("Products")["Sales"].sum()
groupby_products
```

### merge()
Combines two DataFrames based on a common column or index (e.g., merging sales and cost data on "Products").

```python
# Sales DataFrame

Products = ['Computers', 'Phones', 'Shoes']
Sales = [2500, 3000, 1400]

# Creating a DataFrame
df_sales = pd.DataFrame({"Products": Products, "Sales": Sales})
df_sales
```

```python
# Cost DataFrame

Products = ['Computers', 'Phones', 'Shoes']
costs = [1800, 2300, 1000]

# Creating a DataFrame
df_costs = pd.DataFrame({"Products": Products, "Costs": costs})
df_costs
```

Now, if we want to merge the two DataFrames, we can use 
the merge() method. The common column between the two 
DataFrames is the "Products" column. We will merge the 
two DataFrames on this column. 

```python
# Merging the two DataFrames
merged_df = df_sales.merge(df_costs, how='left', on='Products')
merged_df
```

## Selecting Data 

### .loc   
Selects rows and columns by labels (names).

```python
# Creating a DataFrame

names = ['Alice', 'Bob', 'Charlie', 'David']
age = [25, 32, 18, 47]
salary = [50000, 80000, 20000, 120000]

df = pd.DataFrame({"Names": names, "Age": age, "Salary": salary})
df
```

```python
# Select rows by label
first_three_rows = df.loc[1:3]
first_three_rows
```

We can also use the .loc attribute to select specific rows and 
columns from a DataFrame. Lets select the names of Alice and David and their salaries from the DataFrame.

```python
alice_and_david = df.loc[[0,3],['Names','Salary']]
alice_and_david
```

We can also filter the DataFrame by a specific  value in a 
column. Let’s say we want to slice the DataFrame to return 
rows of everyone over 30 years. Here is how we do it: 

```python
over_30 = df.loc[df["Age"] > 30]
over_30
```

### .iloc
Selects rows and columns by index positions.

```python
alice_and_bob = df.iloc[0:2]
alice_and_bob 
```

```python
# Selecting specific rows and columns
alice_and_bob_salaries = df.iloc[:2,[0,2]]
alice_and_bob_salaries
```

##  Pandas Data Visualization Functions 

### plot()  — General Plotting 
Creates visualizations using Matplotlib. Default is a line plot.

**Line Chart** - Shows trends over time or continous data

The below code code merges sales and cost data, then visualizes both as a line plot with products on the x-axis.

```python
import matplotlib.pyplot as plt

# Merging the two DataFrames
merged_df = df_sales.merge(df_costs, how='left', on='Products')

# Plotting a line plot
merged_df.plot(x='Products', y=['Sales', 'Costs'])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```


### Bar Plot
df.plot(kind="bar") → compares values across categories (e.g., sales vs costs).

The code below creates a bar chart comparing sales and costs for each product. 


```python
# Plotting a bar plot
merged_df.plot(kind="bar", x='Products', y=["Sales", "Costs"])

plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```





### Box Plot
df.plot(kind="box") → displays data spread, quartiles, and outliers.

The below code creates a box plot showing the distribution, spread, and outliers for sales and costs across products.


```python
merged_df.plot(kind="box", x="Products", y=["Sales", "Costs"])
plt.ylabel("Sales")
plt.show()

```

### Scatter Plot
df.plot(kind="scatter", x="Sales", y="Costs") → shows relationships between two variables.

The below code creates a scatter plot showing how sales relate to costs for each product.

```python
# Plotting a Scatter plot
merged_df.plot(kind="scatter", x="Costs", y="Sales")
plt.title("Sales vs Costs")
plt.show()
```

### Histogram
df["Sales"].plot(kind="hist") → shows distribution of values in a column.

The below code creates a histogram showing how sales values are distributed across bins. It’s great for spotting skew, spread, or clusters in your data.

```python
# Plotting a hist plot
merged_df["Sales"].plot(kind="hist", bins=5)
plt.title("Sales Distribution")
plt.show()
```

A bin in a histogram is a bucket that groups numbers into ranges, showing how many values fall inside each range.

##  Sorting Data 

### sort_values() 
sort by column values (ascending by default, set ascending=False for descending).



```python
Products = ['Computers', 'Phones', 'Shoes', 'Computers', 'Phones']
Sales = [2500, 3000, 1400, 2100, 2800]

# Creating a DataFrame
df = pd.DataFrame({"Products": Products, "Sales": Sales})

# Sorting the DataFrame by the sales column
df_sorted = df.sort_values(by='Sales')
df_sorted
```

### sort_index() 
sort by row labels (the DataFrame’s index).



```python
Products = ['Computers', 'Phones', 'Shoes', 'Computers', 'Phones']
Sales = [2500, 3000, 1400, 2100, 2800]

# Creating a DataFrame
df = pd.DataFrame({"Products": Products, "Sales": Sales})

# Sorting the DataFrame by the index
df_sorted_by_index = df.sort_index(ascending=False, inplace=False)
df_sorted_by_index
```

### nsmallest() / nlargest()
quickly return the smallest or largest n rows based on a column.

```python
# Creating a DataFrame
df = pd.DataFrame({"Products": Products, "Sales": Sales})

# Getting two products with the least sales
df.nsmallest(2, "Sales")
```
