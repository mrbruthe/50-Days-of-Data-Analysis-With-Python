2. Here is a list below:    
list1 = [2, 3, 4, 6]    
Create a one-dimensional array from the list above. Write 
another line of code to check the shape and size of your 
array. 
Write a function (Python function) that takes this array and 
returns a new array with all the values in the original array 
reversed. 

```python
# Creating one dimentional array
list1 = [2, 3, 4, 6]
arr = np.array(list1)
arr
```

```python
# Checking shape of array
arr.shape
```

```python
# Checking size of array
arr.size
```

arr.shape → structure (rows × columns).

arr.size → total count of elements.

Python Function to reverse array

```python
# Using function to reverse array
def reverse_array (array):
    return array[::-1]

# Creating an array 
arr = np.array(list1)

# Reversing array using function
arr_reversed = reverse_array(arr)
arr_reversed
```

Using np.flip() function to reverse array

```python
arr_flipped = np.flip(arr)
arr_flipped
```

3. list1 = [2, 3, 4, 6]    
list2 = [8, 10, 12, 14]    
Create an array from the two lists above. Then write code to 
check the shape, dimensions,  and data type of the array. 

```python
# Creating an array from the two list
list1 = [2, 3, 4, 6] 
list2 = [8, 10, 12, 14] 

arr = np.array([list1,list2])
arr
```

```python
# to details of the array
arr_shape = arr.shape
arr_dimension = arr.ndim
arr_type = arr.dtype
arr_details = [arr_shape, arr_dimension, arr_type]
arr_details
```

Note : NumPy defaults to int32 on 32‑bit systems and int64 on 64‑bit systems, so the dtype depends on your machine’s architecture.

4. list1 = [2, 3, 4, 6]    
list2 = [8, 10.1, 12, 14]    
list3 = [16, 18, 20, 22.1]    
Create an array from the three (3) lists above. Take note 
that some of the lists contain floats. Create an array that 
will have an int64 data type.

```python
list1 = [2, 3, 4, 6] 
list2 = [8, 10.1, 12, 14]
list3 = [16, 18, 20, 22.1] 

# Creating an array from the lists
arr = np.array([list1, list2, list3])
arr
```

```python
# Convertin the array to int64 data type
int64_arr = arr.astype(np.int64)
int64_arr
```

```python
# print array data type
int64_arr.dtype
```

We can as well specify the data type when the arry is created.

```python
list1 = [2, 3, 4, 6] 
list2 = [8, 10.1, 12, 14]
list3 = [16, 18, 20, 22.1] 

# Creating an array from the lists
arr = np.array([list1, list2, list3], dtype='int64')
arr
```

5. NumPy arrays are homogeneous, so mixed integers and floats become a single type—by default float64.

```python
# list of integers and floats
lst = [1, 2.5, 3, 4, 2, 5]

# Creating an array from a list of floats and integers 
my_array = np.array(lst)
my_array.dtype
```
