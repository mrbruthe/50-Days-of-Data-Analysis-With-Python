1. Import the dataset using pandas. View the first 5 columns.

```python
import pandas as pd
df = pd.read_csv("population_data.csv")
df.head(5)
```

2. Are there any missing values in the dataset? If so, how 
many and in which columns? 


```python
df.isnull().sum()
```

3. Are there any duplicate rows in the "Country" column? 


```python
df.duplicated(subset=["Country"]).sum()
```

4. How many unique countries are there in the "Country" 
column? 


```python
print(f"The number of unique countries are {df["Country"].nunique()}")
```

5. What is the range of the "Population" column values in the 
dataset? 


```python
column_name = 'Population'

# Find the difference between the max value and min value
column_range = df[column_name].max() - df[column_name].min()
print(f'{column_range:,}')
```

6. What are the mean and median of the "Population" column 
values? 


```python
median = df["Population"].median()
mean = df["Population"].mean()

print(f"The median of the population is {median:,} and the mean of the population is {mean:,}")
```

7. Import the population dataset. What is the total population 
of all the countries in the dataset? 


```python
print(f"The total population is {df["Population"].sum():,}")
```

8. What is the average population of all the countries? 


```python
print(f"The average population per country is {df["Population"].mean():,}")
```

9. What is the average urban population of all the countries? 


```python
print(f"The average urban population per country is {df["Urban Population"].mean():,.2f}")
```

10. Which country has the biggest population? 


```python
print(f"The country with the biggest population is {df.groupby('Country')['Population'].sum().idxmax()}")
```

11. Which country has the smallest urban population? 


```python
print(f"The country with the smallest urban population is {df.groupby('Country')['Urban Population'].sum().idxmin()}")
```

12. Using Matplotlib, plot a histogram of the population in the 
dataset. The number of bins for your plot must be 10. Your 
plot must have axis labels and a title. Your plot title's font 
size must be 20. Make your title bold. 

```python
import matplotlib.pyplot as plt 

plt.subplots(figsize=(8, 6))

plt.hist(df["Population"],
        bins=10,
        density=1,
        alpha=0.7,
        edgecolor=None)

plt.ylabel("Frequency")
plt.xlabel("Population")
plt.title('Population Dist', fontsize=12,
            fontweight="bold")
plt.show()
```
