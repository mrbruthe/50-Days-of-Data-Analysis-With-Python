```python
names = ["John", "Kelly", "Jos", "Peter", "Robert", "Piper"]
age = [21, 21, 56, 44, 56, 96]
gender = ['m', 'f', 'm', 'm', 'm', 'f']
```

1. Using the data above, create an array and transpose it. The 
creation of an array and transposition should be combined

```python
import numpy as np
arr = np.array([names, age, gender]).transpose()
arr
```

6. One of the uses of slicing is for data visualization. Slicing is 
often used in data visualization to extract specific portions 
of a dataset for plotting. Write code to create a subarray of 
the ages (only ages) in the transposed array using slicing. 
Using Matplotlib, plot a histogram of the age array. Your 
graph should have an xlabel, an ylabel, and a title. 

```python
age_array = arr[:,1]
age_array
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.title("Age Array")
plt.hist(age_array, rwidth = 2.9, bins=10)
plt.xlabel("age")
plt.ylabel("Frequency")
plt.show
```
